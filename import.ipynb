## import

importの使い方

```python
import numpy as np
#asは別名でのインポート
```

npモジュールのrandom.randnアトリビュートを利用したい場合はnp.の後に続ける

```python
np.random.randn(6,4)
#6行４列のランダムな数値を作成
```

```python
import pandas as pd
```

```python
pd.DataFrame(np.random.randn(6,4))
```

例えばnpオブジェクトに存在するarray関数へのアクセスはnp.arrayでアクセスする。ちょっと確認してみよう

```python
np.array([1,2,3,4])
```

では、これではできないのか？

```python
array([1,2,3,4])
```

そんな関数は定義されていないぞと、怒られます。

ただ、設定さえすれば、関数単位で利用できるようにすることも可能

例えば下記のように

```python
from numpy import array
```

```python
array([1,2,3,4])
#これでarrayという名前で関数が利用できるようになった。
```

全てを指定するのが面倒ということなら

```python
from numpy import *
```

```python
zeros(5)
```

ということも可能

ただし、オブジェクトは上書きされてしまうので、名前が被らないように命名規則などきちんと決めておく必要がある。
